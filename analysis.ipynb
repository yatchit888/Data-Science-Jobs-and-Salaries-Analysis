```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Average Salaries by year

```python
# Grouping data by 'work_year' and calculating average salary
average_salary_by_year = df.groupby('work_year')['salary_in_usd'].mean().reset_index()

# Plotting
plt.figure(figsize=(10, 6))
sns.lineplot(data=average_salary_by_year, x='work_year', y='salary_in_usd')
plt.title('Average Salary by Year')
plt.xlabel('Year')
plt.ylabel('Average Salary (USD)')
plt.show()
```

Frequency of Job Titles

```python
# Counting occurrences of each job title
job_title_frequency = df['job_title'].value_counts().reset_index()
job_title_frequency.columns = ['job_title', 'frequency']

# Plotting
plt.figure(figsize=(12, 8))
sns.barplot(data=job_title_frequency[:10], x='frequency', y='job_title')
plt.title('Frequency of Job Titles')
plt.xlabel('Frequency')
plt.ylabel('Job Title')
plt.show()
```

Top 10 Common Data Jobs

```python
# Selecting the top 10 job titles
top_10_jobs = job_title_frequency.head(10)

# Plotting
plt.figure(figsize=(10, 6))
plt.pie(top_10_jobs['frequency'], labels=top_10_jobs['job_title'], autopct='%1.1f%%')
plt.title('Top 10 Most Common Data Jobs')
plt.show()
```

Salaries by Job Title

```python
# Grouping by 'job_title' and calculating average salary
average_salary_by_title = df.groupby('job_title')['salary_in_usd'].mean().sort_values(ascending=False).reset_index()

# Plotting
plt.figure(figsize=(12, 8))
sns.barplot(data=average_salary_by_title[:10], x='salary_in_usd', y='job_title')
plt.title('Average Salaries by Job Titles')
plt.xlabel('Average Salary (USD)')
plt.ylabel('Job Title')
plt.show()
```

Distribution of Job Categories

```python
# Plotting distribution of job categories
plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='job_category')
plt.title('Distribution of Job Categories')
plt.xlabel('Count')
plt.ylabel('Job Category')
plt.show()
```

Average salaries by Job Category

```python
# Calculating average salary by job category
average_salary_by_category = df.groupby('job_category')['salary_in_usd'].mean().sort_values(ascending=False).reset_index()

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(data=average_salary_by_category, x='salary_in_usd', y='job_category')
plt.title('Average Salaries by Job Category')
plt.xlabel('Average Salary (USD)')
plt.ylabel('Job Category')
plt.show()
```

Salary by Experience level and Employment Type

```python
# Experience Level
average_salary_by_experience = df.groupby('experience_level')['salary_in_usd'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(data=average_salary_by_experience, x='experience_level', y='salary_in_usd')
plt.title('Average Salary by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Average Salary (USD)')
plt.show()
```

Average Salary by Employment Type

```python
# Calculate average salary by employment type
average_salary_by_employment = df.groupby('employment_type')['salary_in_usd'].mean().reset_index()

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(data=average_salary_by_employment, x='employment_type', y='salary_in_usd', palette='coolwarm')
plt.title('Average Salary by Employment Type')
plt.xlabel('Employment Type')
plt.ylabel('Average Salary (USD)')
plt.xticks(rotation=45)  # Rotates the x labels for better visibility if needed
plt.show()
```

Average Salary by Work Setting

```python
# Calculate average salary by work setting
average_salary_by_setting = df.groupby('work_setting')['salary_in_usd'].mean().reset_index()

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(data=average_salary_by_setting, x='work_setting', y='salary_in_usd', palette='viridis')
plt.title('Average Salary by Work Setting')
plt.xlabel('Work Setting')
plt.ylabel('Average Salary (USD)')
plt.xticks(rotation=45)  # Rotates the x labels for better visibility if needed
plt.show()
```

Average Salary by Company Size

```python
# Calculate average salary by company size
average_salary_by_size = df.groupby('company_size')['salary_in_usd'].mean().reset_index()

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(data=average_salary_by_size, x='company_size', y='salary_in_usd', palette='magma')
plt.title('Average Salary by Company Size')
plt.xlabel('Company Size')
plt.ylabel('Average Salary (USD)')
plt.xticks(rotation=45)  # Rotates the x labels for better visibility if needed
plt.show()
```
